```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load data
df = pd.read_csv('C:/Users/USER/Downloads/UNRATE.csv', parse_dates=['observation_date'], index_col='observation_date')

# Enable showing all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Show all rows and columns
print("All Rows and Columns of the DataFrame:")
print(df)

# Descriptive stats
print("\nDescriptive Statistics:")
print(df.describe())

# Plot
df.plot(title='Monthly Unemployment Rate', ylabel='%')
plt.show()

# Decomposition (additive for unemployment data)
result = seasonal_decompose(df, model='additive')
result.plot()
plt.show()
```
